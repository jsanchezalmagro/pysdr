```python
import numpy as np
import matplotlib.pyplot as plt

start_time = 0
end_time = 0.01
sample_rate = 3000
time = np.arange(start_time, end_time, 1/sample_rate)
theta = 0
frequency = 300
amplitude = 1
sinewave = amplitude * np.sin(2 * np.pi * frequency * time + theta)
plt.figure(figsize=(20, 6), dpi=80)
plt.plot(sinewave)
```

```python
a = 0
b = 0.91875
counter = 0
flag = 0
mu = []
acc = 0
rd = []
for i in sinewave:
    if (flag==1 or counter ==1):
        acc = a + b
    a = acc % 1
    if acc < 1:
        flag = 1
    else:
        flag = 0
    if (counter==0):
        counter = 1
    else:
        counter = 0
    mu.append(a)
    rd.append(flag or counter)

plt.figure(figsize=(20, 6), dpi=80)
plt.plot(mu)

a = 0
b = 0.8
counter = 0
flag = 0
mu = []
acc = 0
rd = []
for i in sinewave:
    if (flag==1 or counter ==1):
        acc = a + b
    a = acc % 1
    if acc < 1:
        flag = 1
    else:
        flag = 0
    if (counter==0):
        counter = 1
    else:
        counter = 0
    mu.append(a)
    rd.append(flag or counter)

plt.plot(mu)
```
